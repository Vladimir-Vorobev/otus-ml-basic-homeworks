## Сделайте выводы по анализу датасета и построенным графикам. По желанию можете продолжить анализ данных и также отразить это в выводах.

К сожалению, у датасета нет описания, поэтому я считаю, что это датасет с информацией о посетителях некоего ресторана, где
* total_bill - итоговый чек человека
* tip - некая величина, определяющая размер чаевых
* sex - пол
* smoker - курит ли человек или нет
* day - день недели визита в ресторан
* time - визит в обед или ужин
* size - размер заказа

* В целом из информации выше видно, что интерквартильный размах итогового чека находится в процежутке от 13.3475 до 24.1275, чаевых - от 2 до 3.5625, размера заказа - от 2 до 3
* В выходные дни суммы итогового чека ожидаемо выше, чем в будни, особенно в воскресенье
* В целом мужчины платят и оставляют больше чаевых несколько больше женщин, но не намного (точное подтверждение - (1) и (2))
* В целом чем выше итоговый чек, тем выше чаевые (это подтверждает корреляция ниже)
* В подавляющем большинстве случаев размер заказа 2
* В вечернее время (во время ужина) чаще дают более высокие чаевые
* Чем больше размер заказа, тем больше итоговый чек (4)
* Итоговый чек мужчин-курильщиков больше мужчин-некурильщиков, однако итоговый чек женщин-курильщиков немного меньше женщин-некурильщиков 

```python
new_tips = tips.copy(deep=True)
new_tips['smoker'] = new_tips['smoker'].cat.codes
new_tips['sex'] = new_tips['sex'].cat.codes
new_tips['day'] = new_tips['day'].cat.codes
new_tips['time'] = new_tips['time'].cat.codes

fig = plt.figure(figsize=(8, 8))
corr = new_tips.corr()
corr.style.background_gradient(cmap='coolwarm')
mask = np.zeros_like(corr, dtype=bool)
mask[np.triu_indices(corr.shape[0], k=1)] = True
corr[mask] = np.nan
sns.heatmap(corr, annot=True);
```

Корреляция показывает высокие зависимости между

* time и day, скорее всего данные подобраны неравномерно, т.е. для каждого дня недели подобрано много значений или во время обеда, или во время ужина (3)
* total_bill и tip
* total_bill и size
* tip и size, что означает, что выше чаевые скорее будут от тех, кто заказал больше (6)
